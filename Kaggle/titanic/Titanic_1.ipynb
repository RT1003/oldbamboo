```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data = pd.DataFrame(train_data) 
test_data = pd.DataFrame(test_data) 
```

```python
train_data.head()
```

```python
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())
del train_data['Cabin']
del train_data['Embarked']

test_data['Age'] = test_data['Age'].fillna(test_data['Age'].mean())
del test_data['Cabin']
del test_data['Embarked']
test_data = test_data.dropna()

train_data.isnull().sum()
test_data.isnull().sum()
```

```python
non_null = train_data.copy()
non_null_t = test_data.copy()
```

```python
non_null
```

```python
del non_null['Ticket']
del non_null['Name']

del non_null_t['Ticket']
del non_null_t['Name']
non_null
```

```python
dummies = non_null.copy()
o = pd.get_dummies(dummies['Sex'])

dummies_t = non_null_t.copy()
o_t = pd.get_dummies(dummies_t['Sex'])

o.head()
```

```python
dummies['female'] = o['female']
dummies['male'] = o['male']

dummies_t['female'] = o_t['female']
dummies_t['male'] = o_t['male']
```

```python
del dummies['Sex']

del dummies_t['Sex']

dummies.head()
```

```python
train_Y = np.array(dummies['Survived'])
train_X = np.array(dummies[['Pclass', 'Age', 'SibSp', 'Parch','Fare','female','male']])

test_X = np.array(dummies_t[['Pclass', 'Age', 'SibSp', 'Parch','Fare','female','male']])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,Y_train,Y_valid = train_test_split(train_X,train_Y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape)
print(X_valid.shape)
print(Y_train.shape)
print(Y_valid.shape)
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
svm_model.fit(X_train,Y_train)
Y_pred = svm_model.predict(X_valid)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_valid, Y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(solver='lbfgs')
logistic_model.fit(X_train,Y_train)
Y_pred2 = logistic_model.predict(X_valid)
```

```python
print(classification_report(Y_valid, Y_pred2))
```

```python
svm_model = SVC()
svm_model.fit(X_train,Y_train)
Y_pred3 = svm_model.predict(test_X)
```

```python
logistic_model = LogisticRegression(solver='lbfgs')
logistic_model.fit(X_train,Y_train)
Y_pred4 = logistic_model.predict(test_X)
```
