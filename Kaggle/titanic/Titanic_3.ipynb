```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data = pd.DataFrame(train_data) 
test_data = pd.DataFrame(test_data) 
```

```python
train_data.corr()
```

```python
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())
del train_data['Cabin']
del train_data['Embarked']

test_data['Age'] = test_data['Age'].fillna(test_data['Age'].mean())
del test_data['Cabin']
del test_data['Embarked']
test_data = test_data.fillna(test_data['Fare'].mean())

train_data.isnull().sum()
test_data.isnull().sum()
```

```python
test_data.shape
```

```python
non_null = train_data.copy()
non_null_t = test_data.copy()
```

```python
non_null
```

```python
del non_null['Ticket']
del non_null['Name']

del non_null_t['Ticket']
del non_null_t['Name']
non_null
```

```python
dummies = non_null.copy()
o = pd.get_dummies(dummies['Sex'])

dummies_t = non_null_t.copy()
o_t = pd.get_dummies(dummies_t['Sex'])

o.head()
```

```python
dummies['female'] = o['female']
dummies['male'] = o['male']

dummies_t['female'] = o_t['female']
dummies_t['male'] = o_t['male']
```

```python
del dummies['Sex']
del dummies['PassengerId']

del dummies_t['Sex']
del dummies_t['PassengerId']
```

```python
dummies.head()
```

```python
dummies.corr()
```

```python
dummies.groupby(dummies['Survived']).sum()
```

```python
train_Y = np.array(dummies['Survived'])
train_X = np.array(dummies[['Pclass', 'Age', 'SibSp', 'Parch','Fare','female','male']])

test_X = np.array(dummies_t[['Pclass', 'Age', 'SibSp', 'Parch','Fare','female','male']])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,Y_train,Y_valid = train_test_split(train_X,train_Y, test_size=0.3, random_state=0)
```

```python
print(X_train.shape)
print(X_valid.shape)
print(Y_train.shape)
print(Y_valid.shape)
```

```python
# 数値計算やデータフレーム操作に関するライブラリをインポートする
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
# 図やグラフを図示するためのライブラリをインポートする。
%matplotlib inline
import matplotlib.pyplot as plt
# 機械学習関連のライブラリ群

from sklearn.metrics import confusion_matrix # 混合行列

from sklearn.decomposition import PCA #主成分分析
from sklearn.linear_model import LogisticRegression # ロジスティック回帰
from sklearn.neighbors import KNeighborsClassifier # K近傍法
from sklearn.svm import SVC # サポートベクターマシン
from sklearn.tree import DecisionTreeClassifier # 決定木
from sklearn.ensemble import RandomForestClassifier # ランダムフォレスト
from sklearn.ensemble import AdaBoostClassifier # AdaBoost
from sklearn.naive_bayes import GaussianNB # ナイーブ・ベイズ
```

```python
names = ["Logistic Regression", "Nearest Neighbors", 
         "Linear SVM", "Polynomial SVM", "RBF SVM", "Sigmoid SVM", 
         "Decision Tree","Random Forest", "AdaBoost", "Naive Bayes"]

classifiers = [
    LogisticRegression(),
    KNeighborsClassifier(),
    SVC(kernel="linear"),
    SVC(kernel="poly"),
    SVC(kernel="rbf"),
    SVC(kernel="sigmoid"),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GaussianNB()]
```

```python
from sklearn.metrics import classification_report
for n,c in zip(names, classifiers):
    model = c
    model.fit(X_train,Y_train)
    Y_pred = model.predict(X_valid)
    result = classification_report(Y_pred,Y_valid)
    print(n + '結果は：' + result)
```
